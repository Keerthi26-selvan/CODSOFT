```python
#let us import pandas for manipulating the data
import pandas as pd
#let us import train_test_split to split the for training and testing
from sklearn.model_selection import train_test_split
#let us import the models to predict the output since these problem has many types of models let us import the best two models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

now let us read the csv file

```python
df=pd.read_csv("train.csv")
df
```

Now let us clean the dataframe by removing the null values from the dataframe

```python
df.info()
```

first let us remove the unwanted cols from the datafram

```python
df.drop(['Ticket','Name','Cabin','PassengerId'],axis=1,inplace=True)
```

now the dataset has null value in the age and embarked columns so replace the value with mean of the entire data in columns

```python
df.Age.fillna(df['Age'].mean().round(1),inplace=True)
```

```python
df.Embarked.fillna("U",inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

as now you can see we have removed and fill the unentered columns from the dataframe

now lets map the sex to 1 and 0 because computer can only understand 0s and 1s put male=0 and female=1

```python
df.Sex=df.Sex.map({'male':0 ,"female":1})
```

```python
df
```

now let us remove the label from the dataframe

```python
y=df['Survived'].copy()
X=df.drop(['Survived'],axis=1,inplace=True)
```

```python
df.drop(['Embarked'],axis=1,inplace=True)
```

```python
print(df)
```
